```python
import PyPDF2
from docx import Document
import nltk
document = Document(docx='Фрагмент.docx')
```

```python
text_docx = [par.text for par in document.paragraphs]
```

```python
text_pdf = []
# with open('Фрагмент.pdf', 'rb') as pdf_file:
pdf_reader = PyPDF2.PdfReader('Фрагмент.pdf')
for page_idx in range(len(pdf_reader.pages)):
    pdf_page = pdf_reader.pages[page_idx]
    text_pdf.append(pdf_page.extract_text())
```

```python
text_pdf
```

```python
from nltk import RegexpTokenizer

words_pdf, words_docx = [], []
tokenizer = RegexpTokenizer(r'\w+')
for t_docx in text_docx:

    words_docx.extend([word.lower() for word in nltk.word_tokenize(t_docx, preserve_line=True) if word.isalpha()])
for t_pdf in text_pdf:
    words_pdf.extend([word.lower() for word in nltk.word_tokenize(t_pdf, preserve_line=True) if word.isalpha()])

```

Общее кол-во и кол-во уникальных слов

```python
print(f"Общее кол-во слов в документе docx: {len(words_docx)}")
print(f"Кол-во уникальных слов в документе docx: {len(set(words_docx))}")
print('___________________________________________________')
print(f"Общее кол-во слов в документе pdf: {len(words_pdf)}")
print(f"Кол-во уникальных слов в документе pdf: {len(set(words_pdf))}")
```
